```python
class BinarySearchTree:
    def __init__(self):
        self.tree = {}

    def insert(self, value):
        if not self.tree:
            self.tree[0] = (value, None, None)
            return

        parent_node = None
        current_node = 0
        while current_node is not None:
            parent_node = current_node
            if value < self.tree[current_node][0]:
                current_node = self.tree[current_node][1]
            elif value > self.tree[current_node][0]:
                current_node = self.tree[current_node][2]
            else:
                print("Value already in tree")
                return

        if value < self.tree[parent_node][0]:
            self.tree[parent_node] = (self.tree[parent_node][0], self.tree[parent_node][1], len(self.tree))
            self.tree[len(self.tree)] = (value, None, None)
        else:
            self.tree[parent_node] = (self.tree[parent_node][0], len(self.tree), self.tree[parent_node][2])
            self.tree[len(self.tree)] = (value, None, None)
```
